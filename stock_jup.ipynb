```python
code = 'US.FB'
api.subscribe(code, 'ORDER_BOOK', True)

x = []
y = []

def _handle_broker_queue(param):
    print('Ask:%s~Bid:%s' % (param['Ask'][0], param['Bid'][0]))
    if len(x) == 0:
        x.append(0)
    else:
        x.append(x[-1] + 1)
    y.append(float(param['Ask'][0][0]))
    plt.plot(x, y)
    plt.show()
    clear_output(wait=True)

print(api.get_order_book(code, async_handler=_handle_broker_queue))
api.start()
while 1:
    pass
```
